#### Imports all Neccessary Lib


```python
import os
from openai import OpenAI
from dotenv import load_dotenv
```

#### Load the environment variables (Your token)

```python
load_dotenv() 

token = os.environ.get("GITHUB_TOKEN")
endpoint = "https://models.inference.ai.azure.com"
model_name = "gpt-4o"
```

#### Creating the main function

```python
def main():
    # 2. Initialize the client
    # We point the baseURL to Azure/GitHub, but use the OpenAI library
    client = OpenAI(
        base_url=endpoint,
        api_key=token,
    )

    print("You can start chatting with the assistant. Type 'exit' to quit.")

    # 3. Initialize the conversation history
    # This list acts as the chatbot's "memory"
    messages = [
        {"role": "system", "content": "You are a helpful assistant."}
    ]

    # 4. The Chat Loop
    while True:
        # Get input from the user      
        user_input = input("You: ")

        # Check for exit condition
        if user_input.lower() == "exit":
            print("Goodbye!")
            break

        # Append User's message to history
        messages.append({"role": "user", "content": user_input})

        try:
            # Send the *entire* history to the AI
            response = client.chat.completions.create(
                messages=messages,
                model=model_name
            )

            # Extract the answer
            assistant_message = response.choices[0].message.content
            print(f"Assistant: {assistant_message}")

            # Append AI's message to history so it remembers it next time
            messages.append({"role": "assistant", "content": assistant_message})

        except Exception as err:
            print(f"Error: {err}")
```

#### Calling the Main Function

```python
# Run the main function
if __name__ == "__main__":
    main()
```
